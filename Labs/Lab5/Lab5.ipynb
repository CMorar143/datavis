```r
# Part 8
#Candidate Dataset
first.name <- canddf[c('First.Name')]
surname <- canddf[c('Surname')]
gender <- canddf[c('Gender')]
constituency.n1 <- canddf[c('Constituency.Number')]
count.n1 <- canddf[c('Count.Number')]
party.abb <- canddf[c('Party.Abbreviation')]
party <- canddf[c('Party')]
result <- canddf[c('Result')]
votes <- canddf[c('Votes')]


#Constituency Dataset
constituency.name <- constdf[c('Constituency.Name')]
constituency.n2 <- constdf[c('Constituency.Number')]
count.n2 <- constdf[c('Count.Number')]
cand.number <- constdf[c('Number.Of.Candidates')]
cand.seats <- constdf[c('Number.of.Seats')]
quota <- constdf[c('Quota')]
seats.const <- constdf[c('Seats.in.Constituency')]

#Count Dataset
total <- countdf[c('Total.Votes')]

canddf.data <- data.frame(constituency.n1, first.name, surname, gender, count.n1, party.abb, party, result, votes)
constdf.data <- data.frame(constituency.n2, constituency.name, count.n2, cand.number, cand.seats, quota, seats.const)
countdf.data <- data.frame(total)

#merge(canddf.data, constdf.data, countdf.data)

Reduce(function(x, y) merge(x, y, all=TRUE), list(canddf.data, constdf.data, countdf.data))
```

```r
# Part 9
search <- function(const_name) {
    return(select(canddf[canddf$Constituency == const_name, ][canddf[canddf$Constituency == const_name, ]$Result == 'Elected', ], First.Name, Surname, Constituency, Gender, Party, Party.Abbreviation))
}

search('Louth')
```
