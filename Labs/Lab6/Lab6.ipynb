```r
# Part 6
search <- function(const_name) {
    return(select(canddf[canddf$Constituency == const_name, ][canddf[canddf$Constituency == const_name, ]$Result == 'Elected', ], First.Name, Surname, Constituency, Gender, Party, Party.Abbreviation))
}

search('Louth')
```
